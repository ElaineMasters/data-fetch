```python
#华宝绿色能源 562010
import requests
from bs4 import BeautifulSoup
import csv
from datetime import datetime, timedelta

# 设置请求头
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36 Edg/126.0.0.0"
}

# 输出文件路径
output_file = r'C:\Users\jinwe\Desktop\quant\ETF\data\行业\华宝绿色能源562010ETF .csv'
data_list = []

# 创建CSV文件和写入器
with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    # 写入CSV标题行
    writer.writerow(['日期', '股票代码', '股票名称', '股票数量', '现金替代标志', '现金替代溢价比例', '固定替代金额', '申购溢价比例', '赎回折价比例'])

    start_date = datetime(2024, 1, 1)
    end_date = datetime(2024, 7, 22)

    current_date = start_date
    while current_date <= end_date:
        if current_date.weekday() < 5:  # weekday() 返回星期的索引，星期一为0，星期天为6
            date_str = current_date.strftime('%Y-%m-%d')
            url = 'https://www.fsfund.com/hbdynamic/selDayInfo.do'
            para = {'vc_fundcode': '562010', 'd_tradingday': str(date_str)}

            response = requests.post(url, headers=headers, data=para)
            soup = BeautifulSoup(response.text, 'html.parser')
            table = soup.find_all('table', class_='rates-table')

            # 检查是否找到了表格
            if table:
                table = table[2]
                rows = table.find_all('tr')
                # 检查rows是否为空或者只包含标题行
                if len(rows) > 1:
                    for row in rows[1:]:  # 从第二行开始，跳过标题行
                        columns = row.find_all('td')
                        # 确保列的数量足够
                        if len(columns) >= 7:
                            data = [
                                date_str,
                                columns[0].text.strip(),  # 股票代码
                                columns[1].text.strip(),  # 股票名称
                                columns[2].text.strip(),  # 股票数量
                                columns[3].text.strip(),  # 现金替代标志
                                columns[4].text.strip(),  # 现金替代溢价比例
                                columns[5].text.strip(),  # 申购溢价比例
                                columns[6].text.strip(),  # 赎回折价比例
                            ]
                            data_list.append(data)
                        else:
                            print(f"在 {date_str} 缺少数目不正确的列。")
                else:
                    print(f"在 {date_str} 没有找到数据行。")
            else:
                print(f"在 {date_str} 没有找到表格。")

        current_date += timedelta(days=(1 if current_date.weekday() < 5 else 2))

    # 写入数据到CSV文件
    for row in data_list:
        writer.writerow(row)
```
